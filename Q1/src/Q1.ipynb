## Perceptron on Iris Dataset

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Loading Data

```python
columns = ['sepal_length','sepal_width','petal_length','petal_width','label']
_data = pd.read_csv('../data/iris_train.data',names=columns)
_data_test = pd.read_csv('../data/iris_test.data',names=columns)
```

```python
sns.pairplot(_data, hue="label", palette='colorblind')
```

### Training Perceptron

```python
# Numpy arrays...
data = _data
data_test = _data_test
X = data.ix[:,0:4]
X.insert(0, 'bias', 1)
X = np.array(X)

X_test = data_test.ix[:,0:4]
X_test.insert(0, 'bias', 1)
X_test = np.array(X_test)

y = [1 if x == 'Iris-setosa' else 0 for x in data['label']]
y_test = [1 if x == 'Iris-setosa' else 0 for x in data_test['label']]
```

```python
np.random.seed(0)
alpha = 0.02
m = np.size(X,axis=0)
weights = np.random.rand(5)*2 - 1
iter = 0
while iter<3:
    for i in range(0,m):
        predict = np.dot(X[i,:],weights)
        predict = 1 if predict >= 0 else 0
        if predict != y[i]:
            weights = weights + alpha*(y[i]-predict)*X[i,:].T
    iter += 1
```

```python
def error(X,y,weights):
    predict = np.dot(X,weights)
    predict = [1 if x > 0 else 0 for x in predict]
    errors = [1 if predict[i] != y_test[i] else 0 for i,a in enumerate(predict)]
    print errors
    errors = sum(errors)/float(len(errors))
    return errors
error(X_test,y_test,weights)
```

### Z-scoring

```python
means = _data.mean()
std = _data.std()
data.ix[:,0:4] = (_data.ix[:, 0:4] - means)/std
data_test.ix[:,0:4] = (_data_test.ix[:, 0:4] - means)/std

X = data.ix[:,0:4]
X.insert(0, 'bias', 1)
X = np.array(X)

X_test = data_test.ix[:,0:4]
X_test.insert(0, 'bias', 1)
X_test = np.array(X_test)

y = [1 if x == 'Iris-setosa' else 0 for x in data['label']]
y_test = [1 if x == 'Iris-setosa' else 0 for x in data_test['label']]
```

```python
np.random.seed(0)
alpha = 0.02
m = np.size(X,axis=0)
weights = np.random.rand(5)*2 - 1
iter = 0
while iter<1:
    for i in range(0,m):
        predict = np.dot(X[i,:],weights)
        predict = 1 if predict >= 0 else 0
        if predict != y[i]:
            weights = weights + alpha*(y[i]-predict)*X[i,:].T
    iter += 1
```
