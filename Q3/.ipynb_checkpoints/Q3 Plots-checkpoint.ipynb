### Graphs of results from Q3

```python
import numpy as np
%matplotlib inline
import seaborn as sns
from seaborn import plt
```

```python
def plot(train,test, title=None):
    plt.figure(figsize=(7,7))
    train = np.array(train)
    test = np.array(test)
    if title:
        plt.title("Training and test loss with iterations for "+ title)
    line_up, = plt.plot(train, label='train loss')
    line_down, = plt.plot(test, label='test loss')
    plt.legend(handles=[line_up, line_down])
```
